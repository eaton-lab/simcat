```python
import simcat
import toytree
import numpy as np
import itertools as itt
```

```python
## generate a species tree
tree = toytree.rtree.coaltree(ntips=8, seed=12345)
tree = tree.mod.node_scale_root_height(3)
c, a = tree.draw(tree_style='c')
```

```python
## get observed data w/o migration
model = simcat.Model(
    tree=tree, 
    admixture_edges = (
        (13, 12, None, None, None),
    ),
    theta=(2.0, 0.02),
    nsnps=10000,
    ntests=10,
    nreps=1,
    #debug=True,
    )
```

```python
## init a database of simulations
db1 = simcat.Database(
    name="test-1", 
    workdir="../databases", 
    tree=tree, 
    nedges=1,
    ntests=10,
    nreps=10,
    theta=(0.01, 0.1),
    force=True,
)
```

```python
db1.run(auto=True)
```

```python
from simcat.Database import *
pool = Parallel(db1, rkwargs={}, ipyclient=None, show_cluster=True, auto=True)
pool.wrap_run()
```

```python
pool.tool.ipcluster["cores"] = 4
pool.start_ipcluster()
```

```python
pool.ipyclient = pool.wait_for_connection()
```

```python
pool.print_cluster_info()
```
