## Figures

```python
import simcat
import toytree
import toyplot
import numpy as np
```

```python
# draw the count matrix to use as a reference
simcat.plot.draw_count_matrix();
```

```python
# generate a random 5 tip tree and draw it
tree = toytree.rtree.unittree(5, seed=11111).mod.node_scale_root_height(3)
tree.draw(tree_style='c');
```

# Simulations
Simulate a single data set under a simple no gene flow scenario. Currently we find that many cells in the matrix are not filled. This is because we are using an infinite sites mutation model (only one mutation can occur at any site) thus patterns like CGTA cannot arise. Using a more complex substitution model should add more information to this method. For now, though, the idea holds the same...

```python
# simulate count data 
data = simcat.Model(tree, ntests=1, nsnps=5000, theta=0.5, run=True)

# draw count data for the first matrix
simcat.plot.draw_count_matrix(data.counts[0][0]);
```

### Simulations: introgression (`1 -> 3`)
Simulate three data sets: one with no gene flow, and two with increasing amounts of gene flow. 

```python
# 0% gene flow
data0 = simcat.Model(
    tree, ntests=1, theta=1.0, nsnps=10000, nreps=2, seed=123, run=True)

# 10% gene flow
data1 = simcat.Model(
    tree, ntests=1, theta=1.0, nsnps=10000, nreps=2, seed=123, run=True, 
    admixture_edges=((1, 3, 0.5, 0.5, 0.10),))

# 25% gene flow
data2 = simcat.Model(
    tree, ntests=1, theta=1.0, nsnps=10000, nreps=2, seed=123, run=True, 
    admixture_edges=((1, 3, 0.5, 0.5, 0.25),))
```

### Simulations: introgression (`3 -> 1`)
Simulate three data sets: one with no gene flow, and two with increasing amounts of gene flow. 

```python
# 0% gene flow
data3 = simcat.Model(
    tree, ntests=1, theta=1.0, nsnps=10000, nreps=2, seed=123, run=True)

# 10% gene flow
data4 = simcat.Model(
    tree, ntests=1, theta=1.0, nsnps=10000, nreps=2, seed=123, run=True, 
    admixture_edges=((3, 1, 0.5, 0.5, 0.10),))

# 25% gene flow
data5 = simcat.Model(
    tree, ntests=1, theta=1.0, nsnps=10000, nreps=2, seed=123, run=True, 
    admixture_edges=((3, 1, 0.5, 0.5, 0.25),))
```

### View the SNP counts
It is hard to see differences between simulations from looking only at the filled matrices, since the most common patterns matching the species tree will still tend to look the most common. When we look at *differences* between matrices simulated under different demographic scenarios, however, it is clear that each scenario leaves a characteristic signature. 

```python
# 5 quartet matrices for a scenario with no migration
simcat.plot.draw_five_taxon_matrix(data)
```

```python
# 5 quartet matrices for a scenario with 25% migration
simcat.plot.draw_five_taxon_matrix(data0)
```

### Differences between matrices
When we draw the DIFFERENCE between data sets on each quartet matrix we see that the data with migration is different from the data without migration in a very clear way.  Moreover, the two plots below show the effect of migration (1->3) and migration (3->1) which yield very different patterns, showing that polarization of introgression should be clearly possible from these data. 

Finally, the bottom plot shows a comparison between two data sets that both had no migration. The "noise" in the data caused by stochastic coalescence is much smaller than the signal of introgression, but not tiny. 


```python
# draw differences (1 -> 3)
simcat.plot.draw_five_matrix_comparison(data2, data0, domain_max=100, domain_min=-100)
```

```python
# draw differences (3 -> 1)
simcat.plot.draw_five_matrix_comparison(data5, data3, domain_max=100, domain_min=-100)
```

```python
# draw differences on two simulations both without introgression
simcat.plot.draw_five_matrix_comparison(data0, data3, domain_max=100, domain_min=-100)
```

### Statistics
Because ABBA and BABA are invariants contained in the matrices we can easily calculate introgression based on D-stats for each scenario. As described previously, false positives are expected in cases of phylogenetic non-independence of samples in the four-taxon tests. 

```python
# D-statistics are near zero for no-migration scenario
simcat.utils.abba_baba(data0.counts)
```

```python
# D-statistics are very high for with-migration scenarios
simcat.utils.abba_baba(data2.counts)
```

```python
# Some 4-taxon tests have non-significant D-stats, as expected, like (0, 1, 2, 4)
simcat.utils.abba_baba(data5.counts)
```
